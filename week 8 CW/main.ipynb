```python
import SparseVector as sv

a = sv.SparseVector(10)
a.add_node(2, 5)
a.add_node(5, 3)
a.add_node(8, 7)

b = sv.SparseVector(10)
b.add_node(1, 2)
b.add_node(2, 4)
b.add_node(5, 1)
```

```python
print("a:", end=" ")
a.print()
print("b:", end=" ")
b.print()
```

```python
c = a + b
print("a + b:", end=" ")
c.print()
```

```python
d = a - b
print("a - b:", end=" ")
d.print()
```

```python
dot = a * b
print(f"Dot product: {dot}")
```
